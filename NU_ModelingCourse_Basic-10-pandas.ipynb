#### Example

Let's revisit the pharmacokinetic analysis we conducted previously.

Our dataset is from (<cite>[1]</cite>). 

```python
pk_IV_file = './data/PK_IV_data_full.csv'
```

```python
pk_data_IV = pd.read_csv(pk_IV_file, skiprows=[1])
```

Using `read_csv`, we have imported our values into what pandas (and R) call a *dataframe*.

```python
pk_data_IV
```

Notice that the data are displayed in a nice tabular format. Very large tables will be presented in a different way.

Let's compute some descriptive statistics about these data:

```python
pk_data_IV.describe()
```

We can select given columns in a dataframe by name:

```python
conc = pk_data_IV['Conc']
AUC = pk_data_IV['AUC_fra']

print conc
```

We can also use a 'dotted' notation to refer to a column:

```python
print pk_data_IV.AUMC[2:6].describe()
print '============'
print pk_data_IV.AUMC[2:].max()
```

Along with columns, we can select a range of rows:

```python
row_subset = pk_data_IV[2:5]
```

Once we have data in a dataframe, it is easy to make basic plots:

```python
pk_data_IV.plot(x='Time')
```

```python
pk_data_IV.plot??
```

```python
pk_data_IV.plot(x='Time', y='AUMC', style='go')
```

```python
pk_data_IV.plot(y='AUC_cum', kind='bar')
```

```python
ax1 = pk_data_IV.plot(x='Time', y='AUC_cum')
ax2 = pk_data_IV.plot(x='Time', y='C*T')
```

There are other types of plots that are very useful in statistical analyses, such as *boxplots*,

```python
df = pd.DataFrame(rand(10,5))
bp = df.boxplot()
```

```python
print df.describe()
```

```python
from pandas.tools.plotting import scatter_matrix
sm_data = pd.DataFrame(randn(1000, 3), columns=['A', 'B', 'C'])
sm = scatter_matrix(sm_data, alpha=0.2, figsize=(8, 8), diagonal='kde')
```
