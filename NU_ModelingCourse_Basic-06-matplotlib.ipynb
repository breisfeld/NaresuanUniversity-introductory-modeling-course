**matplotlib** is a python 2D plotting library that produces publication quality figures in a variety of hardcopy formats and interactive environments across platforms. matplotlib can be used in python scripts, the python and ipython shell, web application servers, and six graphical user interface toolkits.

matplotlib tries to make easy things easy and hard things possible. You can generate plots, histograms, power spectra, bar charts, errorcharts, scatterplots, etc, with just a few lines of code.

matplotlib has a number of ways of creating plots. Presented here is one way that provides a lot of flexibility.

Almost every element of a plot in matplotlib can be customized.

Compared to creating plots with Excel and similar tools, matplotlib...

* is more difficult to use at first and requires more practice.
* is much more flexible; every element of a plot is customizable.
* plots are more professional in appearance and can be publication quality.
* plots are set up in code and can be used to produce a consistent *template* for similar plot.

In matplotlib, you can set up a preference file (named `matplotlibrc`) in which you can save your favorite settings (font sizes, tic labels, etc.) and have those used every time.

A user's guide for matplotlib can be found [here](http://matplotlib.org/users/index.html).

## Creating a Scatter Plot with Data

Let's assume that you have simulation results (`x_simp`, `y_simp`) that you would like to plot as a simple connected line.

Let's use the matplotlib function/method `plot` to carry out this task:

```python
x_simp
```

```python
fig_simpleline, axes_simpleline = plt.subplots()
axes_simpleline.plot(x_simp, y_simp, 'ro-')  # a green line
axes_simpleline.set_xlabel('x')
axes_simpleline.set_ylabel('y')
axes_simpleline.set_title('Our great results!')
```

In the above code, the first line sets up the axes and figure we will need.

The names we chose (e.g., `fig_simpleline`, `axes_simpleline`) are not important. We just need to be consistent and use those same names when we use the `plot` function and other related functions.

We now would like to plot our simulation results along with the experimental data (`x_simp_sparse`, `y_simp_sparse`).

We can do this using the same method as before, but with an additional `plot` command.

```python
fig_line_w_pts, axes_line_w_pts = plt.subplots()
axes_line_w_pts.plot(x_simp, y_simp, 'r--')                # a red solid line
axes_line_w_pts.plot(x_simp_sparse, y_simp_sparse, 'go')  # green circles for markers
xlbl = axes_line_w_pts.set_xlabel('x')
ylbl = axes_line_w_pts.set_ylabel('y')
ttl = axes_line_w_pts.set_title('Comparison of Model Results and Data')

plt.savefig('./results/line_plot.png')
```

Now that we know the basics of line and scatter plots, let's get fancy and specify a lot of properties for the lines and markers (plot symbols):

```python
fig_fancyline, axes_fancyline = plt.subplots()
line1 = axes_fancyline.plot(x_simp, y_simp, label=r'$ \alpha=1 $')
line2 = axes_fancyline.plot(x_simp_sparse, y_simp_sparse, label=r'$ \alpha=2 $')
xlbl = axes_fancyline.set_xlabel(r'$\pi \cdot r^2 \cdot \sqrt{y} $')
ylbl = axes_fancyline.set_ylabel('y')
ttl = axes_fancyline.set_title('Fancy Plot')

# now, let's specify how each element will look:
# 'setp' allows us to set up a lot of properties at once
plt.setp(line1, color='g', linewidth=12.0, linestyle='--')
plt.setp(line2, marker='s', linestyle='', markersize=10.0,
         color='#9a679f', markeredgecolor='k', markeredgewidth=3)
plt.setp(xlbl, fontsize=24)
plt.setp(ttl, fontsize=32, family='serif')
ticklabels = plt.setp(axes_fancyline.get_yticklabels(), fontsize=24, color='b')
lgd = axes_fancyline.legend(loc='best')
```

Under many circumstances, it is very useful to put multiple subplots on a single figure.

This allows us to compare different results easily.

We can do this by creating a grid of subplots and plotting in the axis of each of the subplots.

Here is an example with a grid of 2 rows and three columns.

```python
fig_multiplot, axes_multiplot = plt.subplots(nrows=2, ncols=3)
axes_multiplot[0,0].plot(x_simp, y_mult_1, 'r--')
axes_multiplot[0,0].plot(x_simp, 2*y_mult_1, 'k*-')
axes_multiplot[0,1].plot(y_simp, x_simp, 'k*-')
axes_multiplot[0,2].plot(y_simp_sparse, x_simp_sparse, 'c^')
axes_multiplot[1,0].plot(x_simp, y_mult_2, '-', linewidth=5)
axes_multiplot[1,1].plot(y_mult_1, y_mult_2, 'ro')
axes_multiplot[1,2].plot(x_simp, y_simp, ':', linewidth=5)
```

## Creating a Bar Chart

Suppose we have test scores for four different groups of males and females (G1, G2, G3, G4). For each group, we would like to visualize the scores and uncertainties.

To make a bar chart, the method we will use is named `bar`. We can use it as follows:

```python
N = 5
ind = np.arange(N)  # the x locations for the groups
width = 0.35        # the width of the bars
fig_bar, axes_bar = plt.subplots()
rects1 = axes_bar.bar(ind, men_means, width, color='#4B0082', yerr=men_std)
rects2 = axes_bar.bar(ind + width, women_means, width, 
                      color='g', yerr=women_std)
axes_bar.set_ylabel('Scores')
axes_bar.set_title('Scores by Group and Gender')
axes_bar.set_xticks(ind + width)
axes_bar.set_xticklabels( ('G1', 'G2', 'G3', 'G4', 'G5') )
```

## Creating a Histogram

Histograms are often a very useful way to show *distributions* of quantities. 

Suppose we wish to visualize the probability of a certain IQ score in a population. Let's assume that IQ scores follow a normal distribution with a mean of 100 and a standard deviation of 15.

To make a histogram, we will use the method `hist` as follows:

```python
fig_hist, axes_hist = plt.subplots()
n, bins, patches = axes_hist.hist(x_hist, 50, normed=1, 
                                  facecolor='green', alpha=0.75)
axes_hist.set_xlabel('IQ Score')
axes_hist.set_ylabel('Probability')
axes_hist.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')
axes_hist.grid(True)
```

```python
"""Produce custom labelling for a colorbar.

Contributed by Scott Sinclair
"""

import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from numpy.random import randn

# Make plot with vertical (default) colorbar
fig, ax = plt.subplots()

data = np.clip(randn(250, 250), -1, 1)

cax = ax.imshow(data, interpolation='nearest', cmap=cm.coolwarm)
ax.set_title('Gaussian noise with vertical colorbar')

# Add colorbar, make sure to specify tick locations to match desired ticklabels
cbar = fig.colorbar(cax, ticks=[-1, 0, 1])
cbar.ax.set_yticklabels(['< -1', '0', '> 1'])# vertically oriented colorbar

# Make plot with horizontal colorbar
fig, ax = plt.subplots()

data = np.clip(randn(250, 250), -1, 1)

cax = ax.imshow(data, interpolation='nearest', cmap=cm.afmhot)
ax.set_title('Gaussian noise with horizontal colorbar')

cbar = fig.colorbar(cax, ticks=[-1, 0, 1], orientation='horizontal')
cbar.ax.set_xticklabels(['Low', 'Medium', 'High'])# horizontal colorbar

plt.show()
```

```python
"""
Demonstrates similarities between pcolor, pcolormesh, imshow and pcolorfast
for drawing quadrilateral grids.

"""
import matplotlib.pyplot as plt
import numpy as np

# make these smaller to increase the resolution
dx, dy = 0.15, 0.05

# generate 2 2d grids for the x & y bounds
y, x = np.mgrid[slice(-3, 3 + dy, dy),
                slice(-3, 3 + dx, dx)]
z = (1 - x / 2. + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)
# x and y are bounds, so z should be the value *inside* those bounds.
# Therefore, remove the last value from the z array.
z = z[:-1, :-1]
z_min, z_max = -np.abs(z).max(), np.abs(z).max()



plt.subplot(2, 2, 1)
plt.pcolor(x, y, z, cmap='RdBu', vmin=z_min, vmax=z_max)
plt.title('pcolor')
# set the limits of the plot to the limits of the data
plt.axis([x.min(), x.max(), y.min(), y.max()])
plt.colorbar()



plt.subplot(2, 2, 2)
plt.pcolormesh(x, y, z, cmap='RdBu', vmin=z_min, vmax=z_max)
plt.title('pcolormesh')
# set the limits of the plot to the limits of the data
plt.axis([x.min(), x.max(), y.min(), y.max()])
plt.colorbar()



plt.subplot(2, 2, 3)
plt.imshow(z, cmap='RdBu', vmin=z_min, vmax=z_max,
           extent=[x.min(), x.max(), y.min(), y.max()],
           interpolation='nearest', origin='lower')
plt.title('image (interp. nearest)')
plt.colorbar()



ax = plt.subplot(2, 2, 4)
ax.pcolorfast(x, y, z, cmap='RdBu', vmin=z_min, vmax=z_max)
plt.title('pcolorfast')
plt.colorbar()



plt.show()
```
